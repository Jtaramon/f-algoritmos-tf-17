```python
from bokeh.core.properties import value
from bokeh.models import ColumDataSource
from bokeh.transform import dodge

output_file(matrimonios.html)

provincias = ['Loja','guayas']
names = ['hombres','mujeres']

data = {'capital': provinciaas,'Hombres' ; [15151,16151,15561],
       'Mujeres' : [651561,65165165,65156]}

source = ColumnDataSource(data=data)

p = figure(x_range = provincias, y_range = (0,20000000),plot_height=350,title="Matrimonios Ecuador", toolbar_location = None, tools="")
p.vbar(x=dodge('capital', -0.25,range = p.x_range),top='hombres',width=0.2,source=source,
      color="#c9d9d3",legend=value("H"))
p.vbar(x=dodge('capital', -0.25,range = p.x_range),top='mujeres',width=0.2,source=source,
      color="#718dbf",legend=value("M"))

p.x_range.range_padding =0.1
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
